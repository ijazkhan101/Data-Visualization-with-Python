```python
# import libraries

import pandas as pd 
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt 
```

Exmple 1

```python
# tips dataset
tips = sns.load_dataset('tips')
tips.head(12)
```

```python
#scatter plot using matplotlib
# 1. when bothdata are continuous / numberical

plt.scatter(tips['total_bill'],tips['tip'])

#add lables

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter plot using matplotlib')
plt.show()
```

```python
#scatter plot using seaborn
sns.scatterplot(x='total_bill', y='tip', data=tips ,color='red')
```

```python
#scatter plot using matplotlib
# 1. when bothdata are continuous / numberical

plt.scatter(tips['total_bill'],tips['tip'], c=tips['size'])

#add lables

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter plot using matplotlib')
plt.colorbar()
plt.show()
```

```python
#scatter plot using matplotlib
# 1. when bothdata are continuous / numberical

plt.scatter(tips['day'],tips['tip'], c=tips['size'])

#add lables

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter plot using matplotlib')
plt.colorbar()
plt.show()
```

```python
# scatter plot with days agaist tip
plt.plot(tips['day'])
plt.plot(tips['tip'])

plt.xlabel('day')
plt.ylabel('Tip')
plt.title('Scatter plot using matplotlib')
plt.show()
```

```python
# stem plot

plt.stem(tips['day'],color='red')
```

```python
# histogram plot

plt.hist(tips['tip'], bins=15,color='green')
```

```python
#hist plot with seabron

sns.histplot(tips['tip'],bins=15,color='red')
```
