```python
# import libraries

import pandas as pd 
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt 


# data set load 
df= pd.read_excel('data_viz.csv')
df


# getting the data view
```

```python
# check employment_status in the dataset
sns.countplot(x=df['employment_status'])
```

```python
#How to make categorical Plot boxen

sns.catplot(data=df, x='gender' ,y='age', hue="qualification", kind='boxen')
```

```python
#How to make categorical Plot bar

sns.catplot(data=df, x='gender' ,y='age', hue="qualification", kind='bar')
```

```python
#How to make categorical Plot violin

sns.catplot(data=df, x='gender' ,y='age', hue="qualification", kind='violin', color="green", col="gender")
```

```python
#How to make categorical Plot swarm

sns.catplot(data=df, x='gender' ,y='age', hue="employment_status", kind='violin', color="green", col="gender")

sns.swarmplot(data=df, x="gender",y="employment_status", size=3)
```

```python
#boxenplot 

sns.boxenplot(data=df, x='gender',y='age')
```

```python
#boxenplot 

sns.catplot(data=df, x='gender',y='age',kind="box",showmeans=True)
```

```python
sns.catplot(data=df, x="gender",y='salary',kind='bar',capsize=0.1,ci=95,col='qualification')
```
